# Determining the % of Missing Values

```python
## displaying % of columns comprised of nulls
with pd.option_context('display.max_rows', None, 'display.max_columns', None):
    print(Applicant.isnull().sum().sort_values(ascending=False)/len(Applicant)*100)
```

```python
# can see cut off of .7 above
```

## Remove Columns with more than 30% of Misssing values and treating other Missing values

```python
threshold = len(Applicant)* .7
Applicant.dropna(thresh=threshold, axis=1, inplace=True)
```

```python
Applicant['AMT_REQ_CREDIT_BUREAU_DAY'].fillna(value=0, inplace=True)
```

```python
## Replacing Credit Bureau Inquiries null values
Applicant['AMT_REQ_CREDIT_BUREAU_DAY'].fillna(value=0,inplace=True)
Applicant['AMT_REQ_CREDIT_BUREAU_YEAR'].fillna(value=0,inplace=True)
Applicant['AMT_REQ_CREDIT_BUREAU_MON'].fillna(value=0,inplace=True)
Applicant['AMT_REQ_CREDIT_BUREAU_WEEK'].fillna(value=0,inplace=True)
Applicant['AMT_REQ_CREDIT_BUREAU_HOUR'].fillna(value=0,inplace=True)
Applicant['AMT_REQ_CREDIT_BUREAU_QRT'].fillna(value=0,inplace=True)

## Replacing Ext source 3 data w. MEDIAN
Applicant['EXT_SOURCE_3'].fillna(Applicant['EXT_SOURCE_3'].median(),inplace=True)

## Replacing all other rows with missing values
Applicant.dropna(axis=0, inplace=True)
```

```python
# connecting via concat fxn a MODE column and adding it to the table for describe of the ext source 3 column
pd.concat([Applicant['EXT_SOURCE_3'].describe(), pd.DataFrame(Applicant['EXT_SOURCE_3'].mode(), columns=["Mode"]).T])
```

```python
# display 


for col in categorical_col[2:]:
  print(col + ":")
  series_all = Applicant[col].value_counts(normalize = True).mul(100).round(1).astype(str).add('%').rename("All")
  series_1 = Applicant[Applicant['TARGET'] == 1][col].value_counts(normalize = True).mul(100).round(1).astype(str).add('%').rename("1")
  series_0 = Applicant[Applicant['TARGET'] == 0][col].value_counts(normalize = True).mul(100).round(1).astype(str).add('%').rename("0")
  
  print(series_all.to_frame().join(series_1).join(series_0))
  #print(series_1.to_frame())
  print('\n')
```
